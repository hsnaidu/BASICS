Boolean Expression

Combining the Boolean Expression

```python
df[(df['Airline'] == 'IndiGo') &
   (df['Date_of_Journey'] == '24/03/2019')]

# Play with & or || 
```

If you want inverse of this add a ~

```python
df[~(df['Airline'] == 'IndiGo') &
   (df['Date_of_Journey'] == '24/03/2019')]

# Gives me Everything apart from the IndiGo on that date
```

Summarizing My Data

They Basically-use all stat methods

```python
from statistics import mode
print("Min Value")
print(df['Price'].min())
print("Max Value")
print(df['Price'].max())
print("Avg Value")
print(df['Price'].mean())
print("Median Value")
print(df['Price'].median())
print("Mode Value")
print('For this you need import statistics library')
print(mode(df['Price']))
print("Quartile")
print(df['Price'].quantile(0.10))
print("Standard Deviation")
print(df['Price'].std())
print("Variance")
print(df['Price'].var())
print('Count')
print(df['Price'].count())
print("Sum")
print(df['Price'].sum())
```

APPLY

If you want to apply a set of operation or a custom function you can use APPLY

```python
import numpy as np
from statistics import mode

df['Price'].apply([np.mean, np.median, lambda x :mode(x)])
# x is col-- passing col to x on lambda

```

Concat Method

Make a DataFrame using the Query

```python
df.query('Airline == "Jet Airways" or Airline == "IndiGo"').sort_index(axis=0).reset_index(drop=True)
```

```python
pd.concat([df1, df2], ignore_index= True)
# It stacks one below the other 
```

Merging the data

```python
df1 = pd.DataFrame({
    'Index' : [x for x in range(5)],
    'New_index_One' : [x for x in range(10) if x%2 == 0]
})
df2 = pd.DataFrame({
    'Index' : [x for x in range(5)],
    'New_index_Two' : [~x for x in range(10) if x%2 != 0] # Neg becuase I have added ~
})
```

Also adding the suffixes..

```python
df3 = pd.merge(df1, df2, on='Index', how='right', 
               suffixes=['_one','_two']).reset_index(drop=True).drop(columns='Index')

# You can also add SUFFIXES 
```
